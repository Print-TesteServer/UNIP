#1. Introdução a Registros.
Hoje aprenderemos sobre **REGISTROS**, que são Variáveis Compostas **Heterogêneas**, armazenam vários tipos de dados e não somente um como nos vetores e matrizes.

Enquanto **vetores** e **matrizes** nos permitem armazenar vários dados de um **único tipo de dados**, o recurso de **Registro** nos permite armazenar **mais de um tipo de dado**. Um registro é **composto por campos** que especificam cada uma das informações que o compõem. Uma variável do tipo registro é uma **variável composta**, pois engloba um conjunto de dados, e é **heterogênea**, pois **cada campo pode ser de um tipo primitivo diferente**. Muitas coisas do nosso cotidiano se encaixam no conceito de Registro: passagem de ônibus, folha de cheque, livros, filmes, etc. 


#2. Registros em "C"
Uma **struct** é uma variável especial que **contém diversas outras variáveis** normalmente de tipos diferentes.

As **variáveis internas** contidas pela struct são denominadas **membros da struct**.

Podemos dizer que as **structs** da linguagem C são o equivalente ao que se denomina **registros** em outras linguagens de programação.

**Sintaxe:**


```
struct <identificador>
{
 <listagem dos tipos e membros>;
}
struct <identificador> <variavel>;
```

Exemplos: 

1.**Declaração do registro data ( define o formato do registro )**

```
struct data
  {
   int mês;
   int dia;
   int ano;
  };
```
**declaração de variaveis no formato da struct data**

```
  struct data hoje;
  struct dada_data_compra;
  struct data hoje, data_compra;
 ```
  
2.**Declaração do registro livro ( define o formato do registro )**

```
  struct livro {
   char título [40];
   char autor [30];
   int páginas;
   int ano_public;
  }
```
**Declaração da variável redes do tipo struct livro**


```
  struct disciplina redes;
```


# 3. Acesso aos membros do struct

para acessar os membros do struct usa-se o nome da variável seguida de ponto e em seguida o nome no membro.

```
*nome_variável.nome_membro*
```
**Exemplo:**


```
hoje.dia = 10;
data_compra.mes = 2;
redes.ano_public = 2021;
```




**Abaixo podemos ver e executar um programa usando struct (Registro)**

````python
%%writefile reg1.c

#include <stdio.h>

struct data
  {
   int dia;
   char * mes;
   int ano;
  };

int main() {
    
  struct data hoje;
    
  hoje.dia = 25;
  hoje.ano = 2021;
  hoje.mes = "outubro";
  printf( "\n Hoje é %d de %s de %d", hoje.dia, hoje.mes, hoje.ano);
  return(0);
}
````

Compila o arquivo e gera o programa.

````python
%%bash

gcc reg1.c -o reg1
````

Executa o programa.

````python
%%bash
./reg1
````

# 4. Vetor de Registros
Essa é a solução para quando precisamos armazenar vários registros. Para exemplificar o uso desse conceito, vamos supor que tenhamos 5 livros numa biblioteca particular e que nós queremos armazenar as informações desses 5 livros. Como fazer? Simples, faremos um VETOR de REGISTRO. Isso significa que cada um dos 5 posições do Vetor conterá informações de um livro diferente. A Figura  ilustra um vetor de 40 registros.




image.png

**Vamos criar nossa biblioteca e popular a mesma com os títulos.**

````python
%%writefile livros.inp
The_Cathedral_and_the_Bazaar
Eric_Steven_Raymond
132
2011
Python_para_desenvolvedores
Luiz_Eduardo_Borges_
331
2009
Processamento_de_dados_em_larga_escala
Celso_Luiz_Agra
118
2010
Linux_Shortcuts_and_Commands"
Stan_e_Peter_Klimas
212
2020
Linux_Dictionary
Binh_Nguyen
666
2001
````

````python
%%writefile reg2.c

#include <stdio.h>

//Tamanho da biblioteca 
#define tam 5

//definição do registro livro

struct livro {
   char titulo[100];
   char autor[40];
   int paginas;
   int ano_public;
  };

int main() {
  struct livro minha_biblioteca[tam]; // criação do vetor de livros (biblioteca)

   // Le os dados da entrada padrão para as celulas do vetor

  for(int i=0; i < tam ; i++){
      scanf("%s", minha_biblioteca[i].titulo);
      scanf("%s", minha_biblioteca[i].autor);
      scanf("%d", &minha_biblioteca[i].paginas);
      scanf("%d", &minha_biblioteca[i].ano_public);
      }
   printf("\n");
  
  // imprime biblioteca 

  for(int i=0; i < tam ; i++){
      printf("Titulo: %s - ", minha_biblioteca[i].titulo);
      printf("Autor: %s - ", minha_biblioteca[i].autor);
      printf("Paginas: %d - ", minha_biblioteca[i].paginas);
      printf("Ano: %d", minha_biblioteca[i].ano_public);
      printf("\n ");
  }

  return 0;
}
````

Compila o arquivo e gera o programa



````python
%%bash

gcc reg2.c -o reg2
````

Executa o programa passando como entrada padrão o arquivo livros.inp.


````python
%%bash
./reg2 < livros.inp
````
